```python
#this will need to be researched more and explained for understanding,
#but since the genre_ids are in a string representation of a list, 
#instead of a actual list, we want to convert it to an actual list. 
#then, inside of that list, we want to convert all elements to integers
#this is so that each value is a list containing integers, which we can
#later use to convert those integers into our genre mapping and input
#the actual genre names, which will be strings

import ast

# Function to safely convert a string representation of a list to an actual list
def safe_literal_eval(val):
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError):
        return []

# Apply the safe literal evaluation to convert strings to lists
tmdb_movies_df['genre_ids'] = tmdb_movies_df['genre_ids'].apply(safe_literal_eval)

# Function to convert all elements in the list to integers
def convert_genre_ids(genre_ids):
    return [int(genre) for genre in genre_ids]

# Apply the conversion function
tmdb_movies_df['genre_ids'] = tmdb_movies_df['genre_ids'].apply(convert_genre_ids)

# Check the conversion
print(tmdb_movies_df['genre_ids'].head())
```

```python
#in this, we copy our genre_mapping for readability and run code to 
#replace the integers inside our genre_id lists into the actual stringed
#genre name that it represents. 
#this code will need to be researched and explained for understanding

# Define your genre mapping dictionary
genre_mapping = {
    28: "Action",
    12: "Adventure",
    16: "Animation",
    35: "Comedy",
    80: "Crime",
    99: "Documentary",
    18: "Drama",
    10751: "Family",
    14: "Fantasy",
    36: "History",
    27: "Horror",
    10402: "Music",
    9648: "Mystery",
    10749: "Romance",
    878: "Science Fiction",
    10770: "TV Movie",
    53: "Thriller",
    10752: "War",
    37: "Western"
}

# Iterate over genre_ids column and replace integers with genre names
genre_ids = tmdb_movies_df['genre_ids']

for i, genre_id_list in enumerate(genre_ids):
    genre_name_list = []
    for genre_id in genre_id_list:
        genre_name = genre_mapping.get(genre_id)
        if genre_name:
            genre_name_list.append(genre_name)
        else:
            genre_name_list.append(genre_id)  # Keep the original ID if not found in mapping
    tmdb_movies_df.at[i, 'genre_ids'] = genre_name_list

# Check the updated DataFrame
print(tmdb_movies_df['genre_ids'].head())
```

```python
#so now we are going to merge these two dataframes based on
#'title' of the movie. in order to maximize how many columns match up
#we are going to strip the extra whitespace and convert to lowercase
#for the 'title' columns in both dataframes
#we strip the white space and make it lower case and create a new column
#in each dataframe called 'clean_title' and then we merge the dataframes
#off the 'clean_title' column

# Strip whitespace and convert to lowercase for both dataframes
tmdb_movies_df['clean_title'] = tmdb_movies_df['title'].str.strip().str.lower()
movie_budgets_df['clean_title'] = movie_budgets_df['title'].str.strip().str.lower()
merged_df = pd.merge(tmdb_movies_df, movie_budgets_df, on='clean_title', how='inner')
merged_df.info()
```

```python
#this is to preview the new merged_df dataframe and make sure all 
#attributes from each dataframe we are going to use are present,
#including cleaned_title, genre_ids, production_budget, domestic_gross,
#worldwide_gross
merged_df.head()
```

```python
#we figured out that the 'gross' columns are strings, so we need to 
#remove the '$' and the ',' and turn them into integers

merged_df['domestic_gross'] = merged_df['domestic_gross'].str.replace('$', '').str.replace(',', '').astype(int)

merged_df['production_budget'] = merged_df['production_budget'].str.replace('$', '').str.replace(',', '').astype(int)

merged_df['worldwide_gross'] = merged_df['worldwide_gross'].str.replace('$', '').str.replace(',', '').astype(int)

merged_df.head()
```

```python
# now let's remove all rows where domestic gross = 0

merged_df = merged_df[merged_df['domestic_gross'] != 0]
merged_df.head()
```

```python
#now we are going to remove duplicate titles

merged_df = merged_df.drop_duplicates(subset = 'clean_title', keep='first')
merged_df.head()
```

```python
merged_df.info()
```

```python
#now im gonna try to change the date so i can put it in order

#I don't think I have enough columns to only use the last 5 years, so I
#may end up deleting this block

merged_df = merged_df.sort_values(by='release_date_x', ascending=False)
merged_df.head()
```

```python
#let's do an experiment where we only keep movies from the last 10 years
    #and see how many rows we are left with

merged_df = merged_df[merged_df['release_date_x'] >= '2009-01-01']
merged_df.info()
```

```python
#now we are gonna create two more columns to set up our correlation
#calculation.
#this is just to subtract production_budget from domestic_gross
#and worldwide_gross and get our net profit in both domestic and 
#worldwide

merged_df['net_profit_domestic'] = merged_df['domestic_gross'] - merged_df['production_budget']
merged_df['net_profit_worldwide'] = merged_df['worldwide_gross'] - merged_df['production_budget']
merged_df.head()
```

```python
#now we are going to run a pearson correlation to see if there is a 
#correlation between production budget and net profit domestic

correlation_domestic = merged_df['production_budget'].corr(merged_df['net_profit_domestic'])
print(f'The correlation between production budget and domestic net profit: {correlation_domestic}')
```

```python
#now we're gonna run a correlation between production budget and 
#net profit worldwide

correlation_worldwide = merged_df['production_budget'].corr(merged_df['net_profit_worldwide'])
print(f'The correlation between production budget and worldwide net profit: {correlation_worldwide}')
```

```python
#so we can see there is a high correlation between production budget and
#worldwide net profit, but a low correlation between production budget 
#and domestic net profit. we can use this and recommend that if we only
#do domestic movies, we can keep the budget low because there is no
#statistical reason to think it'll create higher net profit, but if we
#do internatioal movies, we should increase our production budget as
#there is a high correlation between budget and worldwide profit

#domestic correlation: 0.1097
#worldwide correlation: 0.672

#in order to see if these are significant, we actually need to run a 
#t-test and find the p-value. then we can compare it to the correlation
#coefficient (r). we will do that now below
```

```python
import numpy as np
from scipy import stats

production_budget = merged_df['production_budget']
net_profit_domestic = merged_df['net_profit_domestic']
net_profit_worldwide = merged_df['net_profit_worldwide']

alpha = 0.05

print("CORRELATION BETWEEN PRODUCTION BUDGET AND DOMESTIC NET PROFIT")

r_domestic, p_value_domestic = stats.pearsonr(production_budget, net_profit_domestic)

print(f"Correlation Coefficient (r): {r_domestic}")
print(f"P-value: {p_value_domestic}")

if p_value_domestic < alpha:
    print("The Correlation between production budget and domestic net profit is statistically significant.")
else:
    print("The Correlation between production budget and domestic net profit is not statistically significant.")
    

print("CORRELATION BETWEEN PRODUCTION BUDGET AND WORLDWIDE NET PROFIT")
r_worldwide, p_value_worldwide = stats.pearsonr(production_budget, net_profit_worldwide)

print(f"Correlation Coefficient (r): {r_worldwide}")
print(f"P-value: {p_value_worldwide}")
      

if p_value_worldwide < alpha:
      print("The correlation bewteen production budget and worldwide net profit is statistically significant.")
else:
      print("The correlation betweenn production budget and worldwide net profit is not statistically significant.")
      
    
```

```python
#since both are statistically significant, we can use our results to 
#make a conclusion. The conclusion is basically there is a low correlation
#between production budget and domestic net profit, so if our movies are
#only going to be released domestically, then production budget is not
#a priority and we can put those expenses towards other things.
#however, if we are releasing the movies worldwide, then we want to 
#put more money towards the production budget because there is a 
#moderate to strong correlation between production budget and 
#worldwide net profit. 
```

```python
#persons table contains person_id, primary_name, and profession(director)
    #writer, actor, etc.

#movie_basics contains movie_id, primary_title

#movie_ratings contains movie_id, averagerating, and numvotes

#directors contains movie_id and person_id

#okay this sql table doesn't contain info about net profit, but what
    #we could do is find the directors with the highest avg ratings

q = """
SELECT p.primary_name AS director_name,
        AVG(averagerating) AS average_rating
    FROM directors d
    JOIN movie_ratings mr ON d.movie_id = mr.movie_id
    JOIN persons p ON d.person_id = p.person_id
    WHERE p.primary_profession LIKE '%director%'
    GROUP BY p.person_id, p.primary_name
    HAVING numvotes > 100000
    ORDER BY average_rating DESC
    LIMIT 10
;
"""
pd.read_sql(q, conn)
```

```python
#now lets do the same query for writers

q = """
SELECT p.primary_name AS writer_name,
        AVG(averagerating) AS average_rating
    FROM writers w
    JOIN movie_ratings mr ON w.movie_id = mr.movie_id
    JOIN persons p ON w.person_id = p.person_id
    WHERE p.primary_profession LIKE '%writer%'
    GROUP BY p.person_id, p.primary_name
    HAVING numvotes > 100000
    ORDER BY average_rating DESC
    LIMIT 10
;
"""
pd.read_sql(q, conn)
```
